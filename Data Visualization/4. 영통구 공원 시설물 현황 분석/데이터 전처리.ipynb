##  영통구 인구 격자 데이터 전처리

```python
import geopandas as gpd
import pandas as pd
from functools import reduce
import fiona
```

### 유아
- 1세 이상 ~ 7세 이하

```python
Child = gpd.read_file('유아.geojson', encoding = 'utf-8')

Child = Child .iloc[:,[2,3]]
Child = Child[Child.val>0]
Child['type'] = '유아'

Child.reset_index(drop = True, inplace = True)
Child.index.name="Seq"

Child
```

```python
Child.to_csv("Child_Pop.csv", encoding='euc-kr')
```

### 초등학생
- 8세 이상 ~ 13세 이하

```python
Element = gpd.read_file('초등학생.geojson', encoding = 'utf-8')

Element = Element.iloc[:,[2,3]]
Element = Element[Element.val>0]
Element['type'] = '초등학생'

Element.reset_index(drop = True, inplace = True)
Element.index.name="Seq"

Element
```

```python
Element.to_csv("Element_Pop.csv", encoding='euc-kr')
```

### 청소년
- 중학생 (14세 이상 ~ 16세 이하) , 고등학생 (17세 이상 ~ 19세 이하)

```python
Middle = gpd.read_file('중학생.geojson', encoding = 'utf-8')
High = gpd.read_file('고등학생.geojson', encoding = 'utf-8')
```

```python
Teenage = pd.merge(Middle, High, how = 'outer', on ='geometry')
Teenage = Teenage.fillna(0)
Teenage['val'] = Teenage['val_x'] + Teenage['val_y']

Teenage = Teenage.iloc[:,[7,3]]
Teenage = Teenage[Teenage.val>0]
Teenage['type'] = '청소년'

Teenage.reset_index(drop = True, inplace = True)
Teenage.index.name="Seq"

Teenage
```

```python
Teenage.to_csv("Teenage_Pop.csv", encoding='euc-kr')
```

### 성인

- 20세 이상 ~ 69세 이하

```python
Twenties = gpd.read_file('20대.geojson', encoding = 'utf-8')
Thirties = gpd.read_file('30대.geojson', encoding = 'utf-8')
Fourties = gpd.read_file('40대.geojson', encoding = 'utf-8')
Fifties = gpd.read_file('50대.geojson', encoding = 'utf-8')
Sixties = gpd.read_file('60대.geojson', encoding = 'utf-8')
```

```python
df = [Twenties, Thirties, Fourties, Fifties, Sixties]
Adult = reduce(lambda left, right : pd.merge(left, right, on ='geometry', how = 'outer'), df)

Adult = Adult.fillna(0)
Adult['val_total'] = Adult.iloc[:,2] + Adult.iloc[:,6] +  Adult.iloc[:,9] +  Adult.iloc[:,12] +  Adult.iloc[:,15]

Adult = Adult.iloc[:,[-1,3]]
Adult = Adult[Adult.val_total>0]
Adult['type'] = '성인'

Adult.reset_index(drop = True, inplace = True)
Adult.index.name="Seq"

Adult
```

```python
Adult.to_csv("Adult_Pop.csv", encoding='euc-kr')
```

### 노인
- 70세 이상 ~ (100세 이상)

```python
Seventies = gpd.read_file('70대.geojson', encoding = 'utf-8')
Eighties = gpd.read_file('80대.geojson', encoding = 'utf-8')
Nineties = gpd.read_file('90대.geojson', encoding = 'utf-8')
Century = gpd.read_file('100세.geojson', encoding = 'utf-8')
```

```python
df2 = [Seventies, Eighties, Nineties, Century]
Old = reduce(lambda left, right : pd.merge(left, right, on ='geometry', how = 'outer'), df2)

Old = Old.fillna(0)
Old['val_total'] = Old.iloc[:,2] + Old.iloc[:,6] + Old.iloc[:,9] + Old.iloc[:,12]

Old = Old.iloc[:,[-1,3]]
Old = Old[Old.val_total>0]
Old['type'] = '노인'

Old.reset_index(drop = True, inplace = True)
Old.index.name="Seq"

Old
```

```python
Old.to_csv("Old_Pop.csv", encoding='euc-kr')
```

### 전체

- 유아 + 초등학생 + 청소년 + 성인 + 노인

```python
total_df = [Child, Element, Teenage, Adult, Old]
total = reduce(lambda left, right : pd.merge(left, right, on ='geometry', how = 'outer'), total_df)

total = total.fillna(0)
total['value_total'] = total.iloc[:,0] + total.iloc[:,3] + total.iloc[:,5] + total.iloc[:,7] + total.iloc[:,9]

total = total.iloc[:,[-1,1]]
total = total[total.value_total>0]
total['type'] = '전체'

total.reset_index(drop = True, inplace = True)
total.index.name="Seq"

total
```

```python
total.to_csv("Total_Pop.csv", encoding='euc-kr')
```

## 영통구 공원 시설물 현황 데이터 전처리

```python
df = pd.read_csv('공원시설목록_파일_정리.csv', encoding ='euc-kr')
```

```python
PARKING = df.iloc[:,0:5]

PARKING = PARKING[PARKING['주차장'] == 1].replace(1,'주차장')
PARKING.rename(columns = {"주차장":"구분"}, inplace = True)
```

```python
SAND = df.iloc[:,[0,1,2,3,5]]

SAND = SAND[SAND['모래놀이'] == 1].replace(1,'모래놀이')
SAND.rename(columns = {"모래놀이":"구분"}, inplace = True)
```

```python
WATER = df.iloc[:,[0,1,2,3,6]]

WATER = WATER[WATER['물놀이'] == 1].replace(1,'물놀이')
WATER.rename(columns = {"물놀이":"구분"}, inplace = True)
```

```python
ZIP = df.iloc[:,[0,1,2,3,7]]

ZIP = ZIP[ZIP['짚라인'] == 1].replace(1,'짚라인')
ZIP.rename(columns = {"짚라인":"구분"}, inplace = True)
```

```python
TRAM = df.iloc[:,[0,1,2,3,8]]

TRAM = TRAM[TRAM['트램플린'] == 1].replace(1,'트램플린')
TRAM.rename(columns = {"트램플린":"구분"}, inplace = True)
```

```python
BASKET = df.iloc[:,[0,1,2,3,9]]

BASKET = BASKET[BASKET['농구장'] == 1].replace(1,'농구장')
BASKET.rename(columns = {"농구장":"구분"}, inplace = True)
```

```python
GAME = df.iloc[:,[0,1,2,3,10]]

GAME = GAME[GAME['게임장'] == 1].replace(1,'게임장')
GAME.rename(columns = {"게임장":"구분"}, inplace = True)
```

```python
FOOT = df.iloc[:,[0,1,2,3,11]]

FOOT = FOOT[FOOT['족구장'] == 1].replace(1,'족구장')
FOOT.rename(columns = {"족구장":"구분"}, inplace = True)
```

```python
SOC = df.iloc[:,[0,1,2,3,12]]

SOC = SOC[SOC['축구장'] == 1].replace(1,'축구장')
SOC.rename(columns = {"축구장":"구분"}, inplace = True)
```

```python
TEN = df.iloc[:,[0,1,2,3,13]]

TEN = TEN[TEN['테니스장'] == 1].replace(1,'테니스장')
TEN.rename(columns = {"테니스장":"구분"}, inplace = True)
```

```python
GATE = df.iloc[:,[0,1,2,3,14]]

GATE = GATE[GATE['게이트볼장'] == 1].replace(1,'게이트볼장')
GATE.rename(columns = {"게이트볼장":"구분"}, inplace = True)
```

```python
REST = df.iloc[:,[0,1,2,3,15]]

REST = REST[REST['휴게쉼터'] == 1].replace(1,'휴게쉼터')
REST.rename(columns = {"휴게쉼터":"구분"}, inplace = True)
```

```python
TOIL = df.iloc[:,[0,1,2,3,16]]

TOIL = TOIL[TOIL['화장실'] == 1].replace(1,'화장실')
TOIL.rename(columns = {"화장실":"구분"}, inplace = True)
```

```python
MULTI = df.iloc[:,[0,1,2,3,17]]

MULTI = MULTI[MULTI['다목적구장'] == 1].replace(1,'다목적구장')
MULTI.rename(columns = {"다목적구장":"구분"}, inplace = True)
```

```python
PARK = pd.concat([PARKING, SAND, WATER, ZIP, TRAM, BASKET, GAME, FOOT, SOC, TEN, GATE, REST, TOIL, MULTI], ignore_index=True)
PARK.index.name="Seq"
PARK
```

```python
PARK.to_csv("PARK.csv", encoding = 'euc-kr')
```
